```python
import numpy as np
import matplotlib.pyplot as plt

# Line to unpack .xvg generated by GROMACS
t_t, temp = np.loadtxt("./temperature.xvg",comments=["@","#"],unpack=True)
t_p, pres = np.loadtxt("./pressure.xvg",comments=["@","#"],unpack=True)
t_e, ene = np.loadtxt("./energy.xvg",comments=["@","#"],unpack=True)
```

```python
plt.figure(dpi=200)
plt.plot(t_t,temp, label="temperature")
plt.scatter(t_t, temp, label="points", color="r")
plt.plot(t_t, np.ones(len(t_t))*300, label="expected")
# plt.plot(x, np.ones(len(x))*y[-1], label=f"reached:{y[-1]:.2f}")
plt.xlabel("time (ps)")
plt.ylabel("K")
plt.title("Evolution of Temperature - MD Run")
plt.legend()
plt.show()
```

```python
plt.figure(dpi=200)
plt.plot(t_p,pres, label="pressure")
plt.scatter(t_p, pres, label="points", color="r")
plt.plot(t_p, np.ones(len(t_p)), label="expected")
# plt.plot(x, np.ones(len(x))*y[-1], label=f"reached:{y[-1]:.2f}")
plt.xlabel("time (ps)")
plt.ylabel("bar")
plt.title("Evolution of Pressure - MD Run")
plt.legend()
plt.show()
```

```python
plt.figure(dpi=200)
plt.plot(t_e,ene, label="energy")
plt.scatter(t_e, ene, label="points", color="r")
plt.xlabel("time (ps)")
plt.ylabel("kJ")
plt.title("Evolution of Energy - MD Run - U + K")
plt.legend()
plt.show()
```

```python
# Gyration Radius
t_g, gyr, _, _, _ = np.loadtxt("./gyrate.xvg",comments=["@","#"],unpack=True)

plt.figure(dpi=200)
plt.plot(t_g, gyr, label="Gyration Radius")
plt.scatter(t_g, gyr, label="points", color="r")
plt.xlabel("time (ps)")
plt.ylabel("nm")
plt.title("Evolution of GR - MD Run")
plt.legend()
plt.show()
```
